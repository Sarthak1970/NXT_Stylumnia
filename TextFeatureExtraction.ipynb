```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english')) - {'in', 'cm', 'kg', 'color', 'size'}

def preprocess_text(text):
    if not isinstance(text, str):
        return ''
    text = text.lower()  # Normalize case
    # Preserve decimal numbers and associated units
    text = re.sub(r'(\d+\.\d+|\d+)(\s?%|\s?in|\s?cm|\s?kg|\s?g|\s?ml|\s?l)', r'\1\2', text)
    # Remove all other unwanted characters except hyphens and underscores
    text = re.sub(r'[^a-zA-Z0-9\s%\-\_\.]', '', text)
    words = text.split()
    # Lemmatize and remove stopwords
    filtered = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return ' '.join(filtered)
```

```python
df_total['description']=df_total['description'].apply(preprocess_text)
df_total['product_name']=df_total['product_name'].apply(preprocess_text)
```

```python
import torch
import torch_xla
import torch_xla.core.xla_model as xm
from transformers import DistilBertTokenizer

# Initialize TPU device
device = xm.xla_device()

# Load the DistilBERT tokenizer
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")

# Convert columns to lists
description_list = df_total['description'].tolist()  # Ensure this is a list of strings
category_name_list = df_total['category_name'].tolist()  # Ensure this is a list of strings

# Tokenize the descriptions
tokens_description = tokenizer(
    description_list,
    max_length=256,
    truncation=True,
    padding=True,
    return_tensors="pt"  # PyTorch tensors
)

# Tokenize the category names
tokens_category_name = tokenizer(
    category_name_list,
    max_length=256,
    truncation=True,
    padding=True,
    return_tensors="pt"  # PyTorch tensors
)

# Move the tokenized tensors to TPU
tokens_description = {key: value.to(device) for key, value in tokens_description.items()}
tokens_category_name = {key: value.to(device) for key, value in tokens_category_name.items()}

# Check tokenized output shapes
print(f"Description Tokens Shape: {tokens_description['input_ids'].shape}")
print(f"Category Name Tokens Shape: {tokens_category_name['input_ids'].shape}")

description_sum = tokens_description['input_ids'].sum(dim=1).to("cpu")  # Bring result back to CPU for printing
category_name_sum = tokens_category_name['input_ids'].sum(dim=1).to("cpu")

print(f"Description Tokens Sum: {description_sum}")
print(f"Category Name Tokens Sum: {category_name_sum}")
```

```python
import torch
from transformers import DistilBertModel
from tqdm import tqdm

model = DistilBertModel.from_pretrained("distilbert-base-uncased").to(device)
embeddings_list = []

# Batch size
batch_size = 128

for start_idx in tqdm(range(0, len(df_total), batch_size), desc="Processing descriptions"):
    end_idx = min(start_idx + batch_size, len(df_total))

    batch_tokens = {key: value[start_idx:end_idx].to(device) for key, value in tokens_description.items()}

    with torch.no_grad():
        description_outputs = model(**batch_tokens)
        description_embeddings = description_outputs.last_hidden_state

    # Use [CLS] token embeddings as the sentence-level representation for each example in the batch
    cls_embeddings = description_embeddings[:, 0, :]  # Shape: [batch_size, hidden_dim]

    # Append the embeddings to the list
    embeddings_list.extend(cls_embeddings.cpu().numpy().tolist())

df_total['distilbert_features'] = embeddings_list

df_total.head()
```
